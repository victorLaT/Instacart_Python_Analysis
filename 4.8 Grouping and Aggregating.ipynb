# Getting the Spending Flag

```python
# Ihad to first create a column for the average prices spent by user
```

```python
df['Average_order'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df
```

```python
# I was then able to create a column with the flags using the loc function.
```

```python
df.loc[df['Average_order'] < 10, 'Spending_flag'] = 'Low Spender'
df.loc[df['Average_order'] >= 10,  'Spending_flag'] = 'High Spender'
```

```python
df
```

```python
df['Spending_flag'].value_counts()
```

# Getting the Frequency Flag

```python
# At first I had change the value'First_order' Because the median wouldnt calculate.
```

```python
df['orders_days_since_order'].replace('First_order',0,inplace=True)
```

```python
# Then I was able to use the group by function to create a column for the median
```

```python
df['median_order'] = df.groupby(['user_id'])['orders_days_since_order'].transform(np.median)
```

```python
# then I had to change the data type.
```

```python
df['orders_days_since_order'] = df['orders_days_since_order'].astype(float)
```

```python
df['orders_days_since_order'].info()
```

```python
df
```

```python
# I was then able to use the loc function to create a column based on the median_order colummn
```

```python
df.loc[df['median_order'] > 20, 'Frequency_flag'] = 'Non-frequent customer'
df.loc[(df['median_order'] > 10) & (df['median_order'] <= 20), 'Frequency_flag'] = 'Regular customer'
df.loc[df['median_order'] <= 10, 'Frequency_flag'] = 'Frequent customer'
```
