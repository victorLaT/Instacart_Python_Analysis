```python
df = pd.read_pickle(os.path.join(path))
```

```python
def Age_Income_type(row):

 if (row['income'] <45000) and (row['Age']<=26):
    return 'Low Income Young  Adult'
 if (row['income'] <45000) and ((row['Age']>26)&(row['Age']<50)):
    return 'Low Income  Adult'
 if (row['income'] <45000)and (row['Age']>=50):
    return 'Low Income Older Adult'
 if (row['income'] >=45000) and (row['Age']<=26):
    return 'High Income Young Adult'
 if (row['income'] >=45000) and ((row['Age']>26)&(row['Age']<50)):
    return 'High Income Adult'
 if (row['income'] >=45000) and (row['Age']>=50):
    return 'High Income Older Adult'    
 
 else: return 'Not enough data'  
```

```python
df['Age_Income_profile'] = df.apply(Age_Income_type, axis =1)
```
